```python
import bs4
import requests
for i in range(1,51):
    genurl="https://books.toscrape.com/catalogue/page-{}.html"
    url=genurl.format(i)
    print(url)
    reqpage=requests.get(url)
    bsoup=bs4.BeautifulSoup(reqpage.text,"lxml")
    products=bsoup.select('.product_pod')
    
    for j in range(20):
        product=products[j]
        if (product.select('.star-rating.Two')!=[]):
            print((product.select('a')[1])['title'])
        else:
            continue
        
        
```

```python
import bs4
import requests
```

```python
res=requests.get('http://quotes.toscrape.com/')
soup1=bs4.BeautifulSoup(res.text,"lxml")
```

```python
htmltag=soup1.select('html')
```

```python
quotes=soup1.select('.quote')
print(quotes)
```

```python
lista=set()
for quote in quotes:
    authortag=quote.select('small')
    lista.add(authortag[0].getText())
    
print('\n'.join(lista))
```

```python
quotes=soup1.select('.quote')
for quote in quotes:
    print(quote.select('.text')[0].getText())
```

```python
tentags=soup1.select('.tag-item')

for tags in tentags:
    tag=(tags.select('.tag'))
    print(tag[0].getText())

```
